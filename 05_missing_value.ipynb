## fillna()

```python
df_copy = df.copy() #재 카피
```

```python
df_copy.tail() #888번 age 빔
```

```python
#df_copy['age'].fillna(100) #콜럼 하나 정해서 그거 찾아주고 값 넣어줌 // 바로 수정이 아니라 연산결과만 반영// 888에 100살 넣음
df_copy['age'] = df_copy['age'].fillna(100)  #카피버전 수정
```

```python
df_copy['deck'].value_counts()
```

```python
df_copy['deck'].fillna('C') #모든 nan이 C로 바뀜
```

```python
df_copy['deck'] = df_copy['deck'].cat.add_categories('Z') #카테고리에 Z추가함
df_copy['deck'].fillna('Z') #카테고리에 지정하지 않은 데이터는 넣을 수 없음 -> 추가해줘야함
```

```python
age_mean = df_copy['age'].mean() #age의 평균값
```

```python
df_copy['age'] = df_copy['age'].fillna(age_mean)
```

```python
df_copy.info()
```

```python
df_copy['age'] = df_copy['age'].fillna(df_copy['age'].median())
```

## practice

- 남자 승객의 결측치는 남자 승객의 평균
- 여자 승객의 결측치는 여자 승객의 평균
- 결측치를 모두 채우고 나서 모든 사람의 평균

```python
df_copy = df.copy()
```

```python
cond1 = df_copy['sex'] == 'male'
```

```python
male_mean = df_copy.loc[cond1, 'age'].mean() #cond1(남자승객)을 고르고 그 나이 정보만 뽑아서 평균치
print(male_mean)
```

```python
cond2 = df_copy['sex'] == 'female'
female_mean = df_copy.loc[cond2, 'age'].mean()
print(female_mean)
```

```python
df_copy.loc[cond1, 'age'] = df_copy.loc[cond1, 'age'].fillna(male_mean) #cond1과 cond2의 평균
df_copy.loc[cond2, 'age'] = df_copy.loc[cond2, 'age'].fillna(female_mean)
```

```python
df_copy['age'].mean()
```
