## where()
-조건이 False인 데이터를 일괄적으로 변환하는

```python
cond = df['fare'] < 30 #30불 보다 적게 낸 사람
df['fare'].where(cond, 1).tail #30불 보다 작은 사람은 1로 출력
```

```python
tips = sns.load_dataset('tips')
# tips.value_counts('day')
# tips.value_counts('tim')
# tips.value_counts('size')
```

- 남성 흡연자 조건 필터링
- tip을 가장 많이 낸 순서로 내림차순 정렬
- 사우이 10개 출력

```python
cond1 = tips['sex'] == 'Male'
cond2 = tips['smoker'] == 'Yes'
tips.loc[cond1 & cond2].sort_values('tip', ascending=False).head(10)
```

- 식사인원이 5명 이상인 저녁테이블
- 흡연자가 없는 주말 테이블
- total_bil, tip 출력

```python
cond1 = tips['size'] >= 5
cond2 = tips['time'] == 'Dinner'

cond3 = tips['smoker'] == 'No'
cond4 = tips['day'].isin(['Sat', 'Sun'])

tips.loc[(cond1 & cond2) & (cond3 & cond4)][['total_bill', 'tip']]
#tips.loc[(cond1 & cond2) & (cond3 & cond4), ['total_bill', 'tip']]
```
